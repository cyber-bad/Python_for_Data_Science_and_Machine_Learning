```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,10)
```

```python
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r')
plt.xlabel('x wala')
plt.ylabel('y wala')
plt.title('cool')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'r--')
```

```python
fig = plt.figure()
```

```python
fig
```

```python
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
```

```python
axes.plot(x, y, 'b')
axes.set_xlabel(' X Label') # Notice the use of set_ to begin methods
axes.set_ylabel(' y Label')
axes.set_title(' Title')
```

```python
fig
```

```python

fig = plt.figure()
axes1 = fig.add_axes([0.9,0.9,0.9,0.9])
axes2 = fig.add_axes([1.0,1.5,0.3,0.2])

axes1.plot(x,y)


axes2.plot(x,y)
```
